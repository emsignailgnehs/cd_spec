```python
from file_load import file_load
import cd_parse2df
import re
import pandas as pd

#input the data folder in 'path', and filename in 'name'
path = r'C:\Users\Sheng\Desktop\CD\20201112_NaClO4_lys10'
name = r'DI_3.txt'

raw_dataset = file_load(path, name)
#print(repr(raw_dataset))
```

```python
rx_dict = {
    'dataset' : re.compile(r'\d*\.?\d*\s\-?\d*.\d*\s\d*.\d*\s\-?\d*.\d*\n'),
}
print(repr(rx_dict['dataset']))
```

```python
dataset = []
for raw_data in raw_dataset:
    temp = re.match(rx_dict['dataset'], raw_data)
    if temp:
        data = temp.group(0).strip('\n').split('\t')
        #print(data)
        dataset.append([float(ind_data) for ind_data in data])
#print(dataset)
df = pd.DataFrame(dataset, columns = ['wavelength (nm)', 'CD (mdeg)', 'HT (V)', 'OD'])
print(df)
```
